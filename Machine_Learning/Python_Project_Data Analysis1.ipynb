# Load & Modify Assignment dataset

```python
# Loading assignment dataset
ass=pd.read_csv('Assignment.csv')

# splittig the dataset into two new column
ass[['name2','name1']]=ass.Name.str.split(',',expand=True)

# Add this two column into one single column again
ass['Aname']=ass['name1'].fillna('')+' '+ass['name2'].fillna('')

#modifying the assignment dataset
ass=ass.iloc[:,[1,6,3]]

# rename the added column from Aname to Full Name
ass.rename(columns={'Aname':'Full Name'},inplace=True)

#make a copy of modified assignment dataset
rs=ass.copy()

# remove extra space from full name column
rs['Full Name'] = rs['Full Name'].str.strip() 

rs.head()
```

# Changing all the NaN value into 0. Using fillna() 

```python
#changing all the NaN values into O
rs=rs.fillna(0)

rs
```

# Calculating attendance

```python
# normal attentance calculation
rs['Att']=(rs['cl 1']+rs['cl 2']+rs['cl 3']+rs['cl 4']+rs['cl 5'])*2

rs
```

# Calculating the quiz marks
**By adding best two.**

```python
#initializing a new column with every row value 0
rs['quiz']=0

# calculation for filling all rows of quiz column
for i in range(0,43,1):
    a=rs['q1 marks'][i]+rs['q2 marks'][i]
    b=rs['q1 marks'][i]+rs['q3 marks'][i]
    c=rs['q2 marks'][i]+rs['q3 marks'][i]
    if (a>b) and (a>c):
        largest=a
    elif (b>a) and (b>c):
        largest=b
    else:
        largest=c
    rs.loc[i,'quiz']=largest

#showing the dataset after adding quiz marks
rs
```

# Calculate Total Marks

```python
'''Total marks calculation by addind Assignment mark,
lab marks,attendance marks and quiz marks
'''

rs['Total']=rs['Ass.']+rs['lab marks']+rs['Att']+rs['quiz']
```

# Calculate Final Grade

```python
#initializing a new column with every row value F
rs['Grade']='F'

# calculation for filling all rows of grade column
for i in range(0,43,1):
    total=rs['Total'][i]
    if total>=90 and total<=100:
        rs.loc[i,'Grade']='A+'
    elif total>=85 and total<90:
        rs.loc[i,'Grade']='A'
    elif total>=80 and total<85:
        rs.loc[i,'Grade']='B+'
    elif total>=75 and total<80:
        rs.loc[i,'Grade']='B'
    elif total>=70 and total<75:
        rs.loc[i,'Grade']='C+'
    elif total>=65 and total<70:
        rs.loc[i,'Grade']='C'
    elif total>=60 and total<65:
        rs.loc[i,'Grade']='D+'   
    elif total>=50 and total<60:
        rs.loc[i,'Grade']='D'
    else:
        rs.loc[i,'Grade']='F'
#showing the dataset after adding grade
rs
```
